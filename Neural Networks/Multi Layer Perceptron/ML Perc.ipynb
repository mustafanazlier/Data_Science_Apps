### Neural network using Sigmoid activation function

```python
import matplotlib.pyplot as plt
import numpy as np

                                            
train_data_sigmoid = np.array(               #inputs for sigmoid
    [
        [0, 0],
        [0, 1],
        [1, 0],
        [1, 1]])
                                            
target_xor_sigmoid = np.array(              #classification of points for sigmoid(0,1) 
    [
        [0],
        [1],
        [1],
        [0]])
```

```python

weights01_Sigmoid = np.array([[5.7, 3.7],  [5.8, 3.7]])  #I have found these weights by running some sample neural network for xor problem
weights12_Sigmoid = np.array([[ 7.5],[-8]])              #then getting the last updated weights from that neural network, then I tried to 
                                                 #simplify/round them, I have to mention that as expected the outputs were far better with the 
                                                 #weights(same weights but with float values like 8.46455152) that obtained with training

# - b01: biases for the  hidden layer
# - b12: bias for the output layer
bias01_Sigmoid = np.array([[-2.4, -5.7]])
bias12_Sigmoid = np.array([[-3.4]])
```

```python
def mySigmoid(x):
    
    return 1 / (1 + np.exp(-x))             #The sigmoid activation function.

def NetworkSigmoid(point):
  
    hiddenLayer1_Sigmoid= np.dot(point, weights01_Sigmoid) + bias01_Sigmoid
    hiddenOut1_Sigmoid = mySigmoid(hiddenLayer1_Sigmoid)
                                                            #My network with sigmoid activation func
    myOutput1_Sigmoid = np.dot(hiddenOut1_Sigmoid, weights12_Sigmoid) + bias12_Sigmoid
    myFinalOutput1_Sigmoid= mySigmoid(myOutput1_Sigmoid)
    
    return myFinalOutput1_Sigmoid

def classifySigmoid(datapoint):
    
     datapoint = np.transpose(datapoint)
     if NetworkSigmoid(datapoint) >= 0.5:
         return 1

     return 0
```

```python
OutputSigmoid = NetworkSigmoid(train_data_sigmoid)

OutputSigmoid
```

```python
plt.figure(figsize=(10, 10))

plt.axis('scaled')
plt.xlim(-5, 5)
plt.ylim(-5, 5)

colors = {
    0: "ro",
    1: "go"
}

# plotting the four datapoints
for i in range(len(train_data_sigmoid)):
    plt.plot([train_data_sigmoid[i][0]],
             [train_data_sigmoid[i][1]],
             colors[target_xor_sigmoid[i][0]],
             markersize=20)

x_range = np.arange(-5, 5, 0.1)
y_range = np.arange(-5, 5, 0.1)

# creating a mesh to plot decision boundary
xx, yy = np.meshgrid(x_range, y_range, indexing='ij')
Z = np.array([[classifySigmoid([x, y]) for x in x_range] for y in y_range])

# using the contourf function to create the plot
plt.contourf(xx, yy, Z, colors=['red', 'green', 'green', 'blue'], alpha=0.4)
```

### Neural network using Tanh activation function 

```python
train_data_tanh = np.array(                    #inputs for tanh
    [                                      
        [-1, -1],
        [-1, 1],
        [1, -1],
        [1, 1]])

target_xor_tanh = np.array(                   #classification of points for tanh(1,-1)
    [
        [-1],
        [1],
        [1],
        [-1]])

                                        
def myTanh(x):
                                             #Tanh activation function.
    return np.tanh(x) 
```

```python

   

weights01_tanh = np.array([[1.6, 1.8], [1.6, 1.8]])  
weights12_tanh = np.array([[-2.9], [ 2.9]])
                                                    #I have obtained these weights by running some sample neural network for xor problem
                                                    #then getting the last updated weights from that neural network, then I tried to
                                                    #simplify/round them, I have to mention that as expected the outputs were far better with the
                                                    #weights(same weights but with float values like 8.46455152) that obtained with training

bias01_tanh = np.array([[-1.4,  1.7]])
bias12_tanh = np.array([[-2.7]])



def NetworkTanh(point):
  
    hiddenLayer1_tanh= np.dot(point, weights01_tanh) + bias01_tanh
    hiddenOut1_tanh = myTanh(hiddenLayer1_tanh)
                                                           #My network with tanh activation func
    myOutput1_tanh = np.dot(hiddenOut1_tanh, weights12_tanh) + bias12_tanh
    myFinalOutput1_tanh= myTanh(myOutput1_tanh)
    
    return myFinalOutput1_tanh


def classifyTanh(datapoint):
    
     datapoint = np.transpose(datapoint)
     if NetworkTanh(datapoint) >= 0:
         return 1

     return -1
```

```python
OutputTanh = NetworkTanh(train_data_tanh)
OutputTanh
```

```python
plt.figure(figsize=(10, 10))

plt.axis('scaled')
plt.xlim(-5, 5)
plt.ylim(-5, 5)

colors = {
    -1: "ro",
    1: "go"
}

# plotting the four datapoints
for i in range(len(train_data_tanh)):
    plt.plot([train_data_tanh[i][0]],
             [train_data_tanh[i][1]],
             colors[target_xor_tanh[i][0]],
             markersize=20)

x_range = np.arange(-5, 5, 0.1)
y_range = np.arange(-5, 5, 0.1)

# creating a mesh to plot decision boundary
xx, yy = np.meshgrid(x_range, y_range, indexing='ij')
Z = np.array([[classifyTanh([x, y]) for x in x_range] for y in y_range])

# using the contourf function to create the plot
plt.contourf(xx, yy, Z, colors=['red', 'green', 'green', 'blue'], alpha=0.4)
```
